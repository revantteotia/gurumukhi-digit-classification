```python
# checking one image from training dataset
path_of_a_two = 'gurumukhi_digits_dataset/train/2/2.bmp'
img = load_img(path_of_a_two)
plt.imshow(img)
```

```python
# checking the size and pixel values of image
print("image size = {}".format(img.size))
print("image mode = {}".format(img.mode))

img_array = img_to_array(img) # converting to np array
print("image array shape = {}".format(img_array.shape))
```

```python
# checking the 3 channels of the image

first_channel = img_array[:,:,0]
second_channel = img_array[:,:,1]
third_channel = img_array[:,:,2]

f, axarr = plt.subplots(1,3)
axarr[0].imshow(first_channel, cmap="gray")
axarr[0].set_title("first channel")
axarr[1].imshow(second_channel, cmap="gray")
axarr[1].set_title("second channel")
axarr[2].imshow(third_channel, cmap="gray")
axarr[2].set_title("third channel")
```

```python
# all 3 channels look same in the image, lets check each pixel value

print(" is first channel equal to second channel", np.array_equal(first_channel, second_channel))
print(" is first channel equal to third channel", np.array_equal(first_channel, third_channel))
```
